```python
x=[1,2,3,4,5]
x2=[2,5,6,7,9]
y1=[8,4,2,6,1]
y2=[2,5,6,3,4]
plt.figure(figsize=(15,6))
plt.plot(x,y1,color='blue',label='Income')
plt.plot(x2,y2,color='red',label='Outcome')
plt.title('distance x time')
plt.xlabel("Distance")
plt.ylabel("Time")
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[8,4,2,6,1]
y2=[2,5,6,3,4]
plt.figure(figsize=(15,6))
plt.plot(x,y1,color='green',label='Income')
plt.plot(x,y2,color='red',label='Outcome')
plt.title('distance x time')
plt.xlabel("Distance")
plt.ylabel("Time")
plt.legend()
plt.show()
```
