# Bar Graph

```python
A=[1,2,3,4,5]
B=[2,4,3,1,9]

plt.bar(A,B,color='blue')

plt.title('population')
plt.xlabel("time period")
plt.ylabel("no of peoples in lakh")

#plt.grid(True,color='blue')
plt.figure(figsize=(10,6))
plt.show()
```

```python
A=[1,2,3,4,5]
B=[2,4,3,1,9]

plt.bar(A,B,color='blue',width=0.5)

plt.title('population')
plt.xlabel("time period")
plt.ylabel("no of peoples in lakh")

#plt.grid(True,color='blue')
plt.figure(figsize=(10,6))
plt.show()
```

```python
A=[1,3,5,7,9]
B=[2,4,3,1,9]
C=[2,4,6,8,10]
D=[5,1,9,3,11]
plt.bar(A,B,label="men",color='blue')
plt.bar(C,D,label="women",color='pink')
plt.title('population')
plt.xlabel("time period")
plt.ylabel("no of peoples in lakh")
plt.legend()
#plt.grid(True,color='blue')
plt.figure(figsize=(15,6))
plt.show()
```

# Pie Chart

```python
# chart your daily activites
slices=[7,2,10,5]
activites=['sleeping','eating','studying','playing']
cols=["purple","red","green","yellow"]
plt.pie(slices,labels=activites,colors=cols)
plt.show()
```

```python
# chart your daily activites
slices=[7,2,10,5]
activites=['sleeping','eating','studying','playing']
cols=["purple","red","green","yellow"]
plt.pie(slices,labels=activites,colors=cols,startangle=200,explode=(0,0.2,0,0),shadow=True,autopct="%1.0f%%%%")
plt.show()
```

# Area Plot / Stack plot

```python
days=[1,2,3,4,5]
eating=[4,3,5,6,4]
sleeping=[8,7,9,6,5]
working=[4,5,8,2,3]
playing=[6,8,6,2,7]
plt.plot([],[],color='purple',label='eating',linewidth=5)
plt.plot([],[],color='red',label='sleeping',linewidth=5)
plt.plot([],[],color='green',label='working',linewidth=5)
plt.plot([],[],color='yellow',label='playing',linewidth=5)
plt.stackplot(days,eating,sleeping,working,playing,colors=['purple','red','green','yellow'])
plt.xlabel('Days')
plt.ylabel('Events')
plt.title("Area plot")
plt.legend()
plt.show()
```
