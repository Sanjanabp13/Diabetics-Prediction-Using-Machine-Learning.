```python
x=[1,2,3,4,5]
y=[9,4,6,5,10]
plt.figure(figsize=(15,6))
plt.plot(x,y,color='#09afeb',linewidth=3)
plt.title('Time and Distance')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.savefig('Mygraph.png')
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[8,4,2,6,1]
y2=[2,5,6,3,4]
plt.figure(figsize=(15,6))
plt.plot(x,y1,color='green',label='Income')
plt.plot(x,y2,color='red',label='Outcome')
plt.title('distance x time')
plt.xlabel("Distance")
plt.ylabel("Time")
plt.legend()
plt.show()
```

```python
A=[1,2,3,4,5]
B=[2,4,3,1,9]

plt.figure(figsize=(10,6))
plt.bar(A,B,color='blue')

plt.title("population")
plt.xlabel("time period")
plt.ylabel("no of peoples in lakh")
plt.grid(True,color='black')

plt.show()
```

```python
A=[1,3,5,7,9]
B=[2,4,3,1,9]
C=[2,4,6,8,10]
D=[5,1,9,3,11]
plt.figure(figsize=(15,6))
plt.bar(A,B,label="men",color='blue')
plt.bar(C,D,label="women",color='pink')
plt.title('population')
plt.xlabel("time period")
plt.ylabel("no of peoples in lakh")
plt.legend()
#plt.grid(True,color='blue')

plt.show()
```

```python
A=[1,2,3,4,5]
B=[2,4,3,1,9]
C=[1,2,3,4,5]
D=[5,1,9,3,11]
plt.figure(figsize=(15,6))
plt.bar(A,B,label="men",color='blue')
plt.bar(C,D,label="women",color='pink',width=0.5)
plt.title('population')
plt.xlabel("time period")
plt.ylabel("no of peoples in lakh")
plt.legend()
#plt.grid(True,color='blue')

plt.show()
```
